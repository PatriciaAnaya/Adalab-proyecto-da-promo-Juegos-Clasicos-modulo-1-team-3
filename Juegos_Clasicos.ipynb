## Piedra Papel Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego continÃºa hasta que un jugador haya acumulado 3 puntos, convirtiÃ©ndose en el ganador.


- Pistas para las alumnas:

    1. Comienza por definir el nÃºmero de puntos objetivo que se deben alcanzar para ganar el juego.

    2. Implementa una funciÃ³n que obtenga la elecciÃ³n del jugador y valide que sea una de las opciones vÃ¡lidas ("Piedra", "Papel" o "Tijera").

    3. Utiliza un bucle `while` para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el nÃºmero de puntos objetivo.

    4. Genera una elecciÃ³n aleatoria para el segundo jugador (la computadora) utilizando el mÃ³dulo `random`.

    5. Determina al ganador de cada ronda y actualiza los puntos correspondientes.

    6. Verifica si algÃºn jugador ha alcanzado el nÃºmero de puntos objetivo y finalicen el juego cuando eso suceda.

```python
import random

def piedra_papel_tijera():
    opciones = ["piedra", "papel", "tijera"]
    resultado = {"puntuacion_j": 0, "puntuacion_r": 0}

    print("Â¡Bienvenida al juego de Piedra ğŸª¨, Papel ğŸ“‹ o Tijera âœ‚ï¸!")
    print("El primero que gane 3 rondas, gana el juego ğŸ…ğŸ†.")

    while True:
        if resultado["puntuacion_j"] >= 3:
            print("ğŸ† Â¡Ganaste el juego! ğŸ†")
            break

        if resultado["puntuacion_r"] >= 3:
            print("Has perdido el juego.ğŸ¥²")
            break

        eleccion_j = input("Indica una de estas tres opciones: 'piedra', 'papel' o 'tijera'. Si quieres parar la partida, indica 'salir'").lower().strip()

        if eleccion_j == 'salir':
            print("Has seleccionado salir. Por abandono, Â¡gana la computadora!")
            break

        if eleccion_j not in opciones:
            print(" âŒ Error: opciÃ³n no vÃ¡lida. Escribe 'piedra', 'papel' o 'tijera'.")
            continue
        
        eleccion_r = random.choice(opciones)
        print(f"TÃº: {eleccion_j} | Rival: {eleccion_r}")

        if eleccion_j == eleccion_r:
            print("Empate. ğŸ¤¯")

        elif (eleccion_j == "piedra" and eleccion_r == "tijera") or \
             (eleccion_j == "papel" and eleccion_r == "piedra") or \
             (eleccion_j == "tijera" and eleccion_r == "papel"):
            resultado["puntuacion_j"] += 1
            print(f"Â¡Ganaste esta ronda! ğŸ‰ \nResultado puntuacion jugador: {resultado['puntuacion_j']} \nResultado puntuacion rival: {resultado['puntuacion_r']}")
        else:
            resultado["puntuacion_r"] += 1
            print(f"Perdiste esta ronda. âŒ \nResultado puntuacion jugador: {resultado['puntuacion_j']} \nResultado puntuacion rival: {resultado['puntuacion_r']}")

    print("Fin del juego. Â¡Gracias por jugar! ğŸ˜Š")
```

 # Ahorcado

- Reglas del juego del "Ahorcado":

    - Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.


- Pistas para las alumnas:

    1. Empieza por seleccionar una palabra secreta para que el jugador adivinador intente adivinar. Puedes crear una lista de palabras.

    2. Inicializa el tablero de juego con guiones bajos ("_") para representar las letras no adivinadas.

    4. Verifica si la letra adivinada por el jugador estÃ¡ en la palabra secreta y actualicen el tablero en consecuencia.

    5. Lleva un seguimiento de las letras adivinadas y los intentos fallidos.

    6. Determina las condiciones de victoria y derrota del juego, y salgan del bucle cuando se cumplan estas condiciones.

```python
def hangman_game():
    limite_intentos = 7
    print(f"Â¿Binvenido al juego del Ahorcado! Tienes {limite_intentos} oportunidades para adivinar la palabra.")
    jugadora1 = input("Jugador 1, introduce tu nombre").strip().capitalize()
    print(f"Â¿Bienvenido, {jugadora1}!ğŸ¤­")

    palabra_resuelta = input(f"{jugadora1}, escoge una palabra a adivinar por jugador 2").strip().upper()
    caracteres_ahorcado = list(palabra_resuelta)
    intentos = 0
    letras_usadas = []
    palabra_escondida = ['_'] * len(caracteres_ahorcado)
    
    print(f"\nLa palabra escondida es:\n{palabra_escondida}")
    print(f"Te quedan {limite_intentos-intentos} intentos.{'\n' *2}")
    jugadora2 = input("Jugadora 2, introduce tu nombre").strip().capitalize()
    print(f"Â¿Bienvenida, {jugadora2}!ğŸ¤­")
    
    while True:

        if "_" not in palabra_escondida:
            print(f"Â¡Enhorabuena {jugadora2}! Has adivinado la palabra escondida: {palabra_resuelta} ğŸ‰ğŸŠ")
            break
    
        if intentos == limite_intentos:
            print(f"Has consumido el mÃ¡ximo de intentos, {jugadora2}!âŒ")
            print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                               / \  |
                                    |
                                =========""") 
            print(f"Â¡{jugadora1} ha ganado el juego!ğŸ‰ğŸŠ")
            break

        if intentos < limite_intentos:
            letra = input(f"{jugadora2} escribe una letra para averiguar si se encuentra en la palabra. Si deseas salir, escribe 'salir'").upper().strip()
            print(f"Las letras que has usado anteriormente son:{sorted(letras_usadas)}")
            print(f"Te quedan {limite_intentos-intentos} intentos para adivinar la palabra. Â¡Suerte! ğŸ€{'\n'*2}")

            if letra == "SALIR":
                print(f"Por abandono, Â¡gana {jugadora1}!")
                print("Esperamos volver a veros pronto, Â¡adiÃ³s! ğŸ˜Š")
                break

            elif len(letra) != 1 or not letra.isalpha():
                print(" Por favor, ingresa solo *una* letra vÃ¡lida (a-z).")
                continue
                
            
            elif letra not in caracteres_ahorcado:
                intentos += 1
                letras_usadas.append(letra)
                print(f"Las letras que has usado anteriormente son:{letras_usadas}")
                print(f"Â¡Incorrecto! ğŸ˜¢ {letra} no se encuentra en la palabra escondida. \nHas consumido {intentos} intento(s) de {limite_intentos}.")
                print("\n" * 2)
                if limite_intentos-intentos == 6:
                    print("""\n
                                +---+
                                |   |
                                    |
                                    |
                                    |
                                    |
                                =========""")
                elif limite_intentos-intentos == 5:
                    print("""\n
                                +---+
                                |   |
                                O   |
                                    |
                                    |
                                    |
                                =========""")                
                elif limite_intentos-intentos == 4:
                    print("""\n
                                +---+
                                |   |
                                O   |
                                |   |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 3:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|   |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 2:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 1:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                               /    |
                                    |
                                =========""") 

            elif letra in caracteres_ahorcado:  
                for elemento in range(len(caracteres_ahorcado)):
                    if caracteres_ahorcado[elemento] == letra:
                        palabra_escondida[elemento] = letra
                letras_usadas.append(letra)
            print(palabra_escondida)
```

# Tic Tac Toe

- Reglas del Juego "Tic-Tac-Toe":

    - El juego se juega en un tablero de 3x3.

    - Dos jugadores, uno representa "X" y el otro "O", alternan turnos.

    - El objetivo es obtener tres sÃ­mbolos del mismo tipo en lÃ­nea horizontal, vertical o diagonal.

    - El juego termina cuando un jugador gana o el tablero se llena (empate).

    - El jugador que logre colocar tres de sus sÃ­mbolos en lÃ­nea es el ganador.


- Pistas para las alumnas:

    1. Comienza creando una representaciÃ³n del tablero de juego utilizando una lista de listas para mantener un seguimiento de las jugadas.

    2. Implementa una funciÃ³n para imprimir el tablero en la pantalla en un formato claro y legible.

    3. Utiliza un bucle `while` para alternar los turnos de los jugadores y recopilar sus movimientos.

    4. Verifica despuÃ©s de cada movimiento si hay un ganador o si el juego termina en empate.

    5. Implementa una funciÃ³n que verifique si un jugador ha ganado alineando tres sÃ­mbolos en fila, columna o diagonal.

    6. Asegurate de validar que una casilla seleccionada por un jugador no estÃ© ocupada antes de permitir el movimiento.

```python
def Tic_tac_toe():

    lista = [" "] * 9
    intentos = 0

    posiciones = {
        "a1": 0, "a2": 1, "a3": 2,
        "b1": 3, "b2": 4, "b3": 5,
        "c1": 6, "c2": 7, "c3": 8
    }

    print("Â¿Bienvenidas a Tic-Tac-Toe! Este juego es muy sencillo, debÃ©is rellenar las casillas disponibles con âŒ o con â­•")
    print("El objetivo es que vuestro sÃ­mbolo haga una lÃ­nea horizontal, vertical o diagonal!")
    print("La jugadora que logre colocar antes sus sÃ­mbolos en lÃ­nea, Â¡gana la partida!")
    print("ğŸ€ğŸ€ Â¡Â¡Â¡Mucha suerte!!! ğŸ€ğŸ€")

    jugadora1 = input("Jugadora 1, por favor, ingresa tu nombre ğŸ˜Š")
    print(f"Bienvenida, {jugadora1}âœ¨")
    jugadora2 = input("Jugadora 2, por favor, ingresa tu nombre ğŸ˜Š")
    print(f"Bienvenida, {jugadora2}âœ¨")

    def revisar_victoria(jugador):
        
        combinaciones = [
            [0,1,2], [3,4,5], [6,7,8],  
            [0,3,6], [1,4,7], [2,5,8],  
            [0,4,8], [2,4,6]           
        ]
        for combo in combinaciones:
            if lista[combo[0]] == lista[combo[1]] == lista[combo[2]] == jugador:
                return True
        return False

    print("Para marcar tu posiciÃ³n usa las coordenadas a,b,c y 1,2,3 Â¡como en un tablero de ajedrez!")

    print(f"a|[{lista[0]}]|[{lista[1]}]|[{lista[2]}]|")
    print(f"b|[{lista[3]}]|[{lista[4]}]|[{lista[5]}]|")
    print(f"c|[{lista[6]}]|[{lista[7]}]|[{lista[8]}]|")
    print(f"   1  2  3")
    
    while intentos < 9:

        
        while True:
            posiciÃ³n_1 = input(f"{jugadora1}, elige quÃ© posiciÃ³n deseas marcar ").lower()
            if posiciÃ³n_1 in posiciones and lista[posiciones[posiciÃ³n_1]] == " ":
                lista[posiciones[posiciÃ³n_1]] = "X"
                intentos += 1
                break

            else:
                print("PosiciÃ³n invÃ¡lida u ocupada, intÃ©ntalo de nuevo. ğŸ¤”")

        print(f"a|[{lista[0]}]|[{lista[1]}]|[{lista[2]}]|")
        print(f"b|[{lista[3]}]|[{lista[4]}]|[{lista[5]}]|")
        print(f"c|[{lista[6]}]|[{lista[7]}]|[{lista[8]}]|")
        print(f"   1  2  3")
        
        if revisar_victoria("X"):
            print(f"Â¡Enhorabuena, {jugadora1}! Has ganado la partida! ğŸ‰ğŸŠ")
            return

        if intentos == 9:
            break

        while True:
            posiciÃ³n_2 = input(f"{jugadora2}, elige quÃ© posiciÃ³n deseas marcar").lower()
            if posiciÃ³n_2 in posiciones and lista[posiciones[posiciÃ³n_2]] == " ":
                lista[posiciones[posiciÃ³n_2]] = "O"
                intentos += 1
                break

            else:
                print("PosiciÃ³n invÃ¡lida u ocupada, intenta de nuevo.")
        
        print(f"a|[{lista[0]}]|[{lista[1]}]|[{lista[2]}]|")
        print(f"b|[{lista[3]}]|[{lista[4]}]|[{lista[5]}]|")
        print(f"c|[{lista[6]}]|[{lista[7]}]|[{lista[8]}]|")
        print(f"   1  2  3")
       
        if revisar_victoria("O"):
            print(f"Â¡Enhorabuena, {jugadora2}! Has ganado la partida! ğŸ‰ğŸŠ")
            return

    print("Â¡Tablas! No hay ganadora. ğŸ™…ğŸ»â€â™€ï¸ğŸ™…ğŸ»â€â™€ï¸")
```
