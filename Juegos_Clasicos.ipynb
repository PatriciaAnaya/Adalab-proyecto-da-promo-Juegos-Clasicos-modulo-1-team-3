## Piedra Papel Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego contin√∫a hasta que un jugador haya acumulado 3 puntos, convirti√©ndose en el ganador.


- Pistas para las alumnas:

    1. Comienza por definir el n√∫mero de puntos objetivo que se deben alcanzar para ganar el juego.

    2. Implementa una funci√≥n que obtenga la elecci√≥n del jugador y valide que sea una de las opciones v√°lidas ("Piedra", "Papel" o "Tijera").

    3. Utiliza un bucle `while` para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el n√∫mero de puntos objetivo.

    4. Genera una elecci√≥n aleatoria para el segundo jugador (la computadora) utilizando el m√≥dulo `random`.

    5. Determina al ganador de cada ronda y actualiza los puntos correspondientes.

    6. Verifica si alg√∫n jugador ha alcanzado el n√∫mero de puntos objetivo y finalicen el juego cuando eso suceda.

```python
import random

def piedra_papel_tijera():
    opciones = ["piedra", "papel", "tijera"]
    resultado = {"puntuacion_j": 0, "puntuacion_r": 0}

    print("¬°Bienvenido al juego de Piedra ü™®, Papel üìã o Tijera ‚úÇÔ∏è!")
    print("El primero que gane 3 rondas, gana el juego üèÖüèÜ.")

    while True:
        if resultado["puntuacion_j"] >= 3:
            print("üéÜ ¬°Ganaste el juego! üéÜ")
            break

        if resultado["puntuacion_r"] >= 3:
            print("Has perdido el juego.ü•≤")
            break

        eleccion_j = input("Indica una de estas tres opciones: 'piedra', 'papel' o 'tijera'. Si quieres parar la partida indica 'salir'").lower().strip()

        if eleccion_j == 'salir':
            print("Has seleccionado salir")
            break

        if eleccion_j not in opciones:
            print(" ‚ùå Error: opci√≥n no v√°lida. Escribe piedra, papel o tijera.")
            continue
        
        eleccion_r = random.choice(opciones)
        print(f"T√∫: {eleccion_j} | Rival: {eleccion_r}")

        if eleccion_j == eleccion_r:
            print("Empate. ü§Ø")

        elif (eleccion_j == "piedra" and eleccion_r == "tijera") or \
             (eleccion_j == "papel" and eleccion_r == "piedra") or \
             (eleccion_j == "tijera" and eleccion_r == "papel"):
            resultado["puntuacion_j"] += 1
            print(f"¬°Ganaste esta ronda! \nResultado puntuacion jugador: {resultado['puntuacion_j']} \nResultado puntuacion rival: {resultado['puntuacion_r']}")
        else:
            resultado["puntuacion_r"] += 1
            print(f"Perdiste esta ronda. \nResultado puntuacion jugador: {resultado['puntuacion_j']} \nResultado puntuacion rival: {resultado['puntuacion_r']}")

    print("Fin del juego. ¬°Gracias por jugar!")
```

```python
piedra_papel_tijera()
```

 # Ahorcado

- Reglas del juego del "Ahorcado":

    - Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.


- Pistas para las alumnas:

    1. Empieza por seleccionar una palabra secreta para que el jugador adivinador intente adivinar. Puedes crear una lista de palabras.

    2. Inicializa el tablero de juego con guiones bajos ("_") para representar las letras no adivinadas.

    3. Utiliza un bucle `while` para permitir que el jugador adivinador haga m√∫ltiples intentos hasta que adivine la palabra o se agoten los intentos.

    4. Verifica si la letra adivinada por el jugador est√° en la palabra secreta y actualicen el tablero en consecuencia.

    5. Lleva un seguimiento de las letras adivinadas y los intentos fallidos.

    6. Determina las condiciones de victoria y derrota del juego, y salgan del bucle cuando se cumplan estas condiciones.

```python
## En construcci√≥n üößüöß

#Ahorcado 
def hangman_game():
    limite_intentos = 7
    print(f"¬øBinvenido al juego del Ahorcado! Tienes {limite_intentos} oportunidades para adivinar la palabra.")
    palabra_resuelta = input("Jugador 1, escoge una palabra a adivinar por jugador 2").strip().upper()
    caracteres_ahorcado = list(palabra_resuelta)
    intentos = 0
    letras_usadas = []
    palabra_escondida = ['_'] * len(caracteres_ahorcado)
    
    print(f"\nLa palabra escondida es:\n{palabra_escondida}")
    print(f"Te quedan {limite_intentos-intentos} intentos.{'\n' *2}")
    
    while True:

        if "_" not in palabra_escondida:
            print(f"¬°Enhorabuena! Has adivinado la palabra escondida: {palabra_resuelta}")
            break
    
        if intentos == limite_intentos:
            print("Has consumido el m√°ximo de intentos! Has perdido el juego")
            print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                               / \  |
                                    |
                                =========""") 
            break

        if intentos < limite_intentos:
            letra = input("Jugador 2, escribe una letra para averiguar si se encuentra en la palabra. Si deseas salir, escribe 'salir'").upper().strip()
            print(f"Las letras que has usado anteriormente son:{sorted(letras_usadas)}")
            print(f"Te quedan {limite_intentos-intentos} intentos para adivinar la palabra. ¬°Suerte! {'\n'*2}")

            if letra == "SALIR":
                print("¬°Gracias por jugar! ¬°Adios!")
                break

            elif len(letra) != 1 or not letra.isalpha():
                print(" Por favor, ingresa UNA letra v√°lida (a-z).")
                continue
                
            
            elif letra not in caracteres_ahorcado:
                intentos += 1
                letras_usadas.append(letra)
                print(f"Las letras que has usado anteriormente son:{letras_usadas}")
                print(f"¬°Incorrecto! {letra} no se encuentra en la palabra escondida. \nHas consumido {intentos} intento(s) de {limite_intentos}.")
                print("\n" * 2)
                if limite_intentos-intentos == 6:
                    print("""\n
                                +---+
                                |   |
                                    |
                                    |
                                    |
                                    |
                                =========""")
                elif limite_intentos-intentos == 5:
                    print("""\n
                                +---+
                                |   |
                                O   |
                                    |
                                    |
                                    |
                                =========""")                
                elif limite_intentos-intentos == 4:
                    print("""\n
                                +---+
                                |   |
                                O   |
                                |   |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 3:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|   |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 2:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                                    |
                                    |
                                =========""") 
                elif limite_intentos-intentos == 1:
                    print("""\n
                                +---+
                                |   |
                                O   |
                               /|\  |
                               /    |
                                    |
                                =========""") 

            elif letra in caracteres_ahorcado:  
                for elemento in range(len(caracteres_ahorcado)):
                    if caracteres_ahorcado[elemento] == letra:
                        palabra_escondida[elemento] = letra
            print(palabra_escondida)
```

    4. Dentro del bucle, presenta la pregunta al jugador y permita que ingresen sus respuestas.

    5. Lleva un seguimiento del n√∫mero de preguntas realizadas y el n√∫mero de respuestas correctas.

    6. Determina las condiciones de victoria y derrota del juego y sal del bucle cuando se cumplan estas condiciones.

```python
preguntasyrespuestas =  {
    "¬øCu√°l es el pa√≠s m√°s peque√±o del mundo?" : "Ciudad del Vaticano",
    "¬øCu√°l es la capital de Albania?" : "Tirana",
    "¬øCu√°ntos oc√©anos hay en la Tierra?" : "5",
    "¬øQu√© pa√≠s tiene m√°s habitantes?" : "China",
    "¬øQu√© r√≠o pasa por m√°s pa√≠ses?" : "Danubio",
    "¬øCu√°l es la ciudad m√°s poblada del mundo?" : "Tokyo",
    "¬øCu√°ntos estados forman parte de EEUU?" : "50",
    "¬øCu√°ntas comunidades aut√≥nomas tiene Espa√±a?" : "17",
    "¬øQu√© continente es el m√°s grande?" : "Asia"
}
import random
def juegopreguntas(preguntasyrespuestas):

        
    contador_aciertos = 0  
    contador_errores = 0

    print(f"¬°Bienvenido al juego de 'Preguntas y Respuestas'! Te realizar√© unas simples preguntas de Geograf√≠a a lo que tendr√°s que responder.\n Para ganar, deber√°s contestar correctamente a 5 preguntas.\n Perder√°s si respondes incorrectamente 3 preguntas.")
    inicio = input("Para continuar, pulsa Intro o introduce 'Salir' para finalizar el juego.")

    if inicio == None:
            True

                    
    if inicio == 'Salir':
            False
                    

    while True:
        pregunta = random.choice(list(preguntasyrespuestas.keys()))
        print(pregunta)
        
                        
        respuesta = input("Por favor, introduce tu respuesta, si quieres parar el juego introdice: 'Salir'").strip().capitalize()
       
        
        if respuesta == 'Salir':
            print("Gracias por participar")
            break 
        
        print(f"Tu respuesta ha sido: {respuesta}")                    
             
        if respuesta == preguntasyrespuestas[pregunta]:
            contador_aciertos += 1
            preguntasyrespuestas.pop(pregunta)
            print(f"Has acertado la pregunta, llevas {contador_aciertos} preguntas acertadas")
           
        
        else:                  
            contador_errores +=1
            print(f"La respuesta es incorrecta, llevas {contador_errores} fallos")
            
           
        
    
        if contador_errores == 3: 
                    print("Has tenido 3 fallos, has perdido el juego")
                    break
                    
        if contador_aciertos == 5:
                    print("Has acertado 5 respuestas, has ganado el juego")
                    break
```

```python
juegopreguntas(preguntasyrespuestas)
```
